```python
import seaborn as sns
import numpy as np
```

```python
def _lr(x, init=0.01, decay=0.9, steps=50):
    return init * decay**(x / steps)

exp = np.vectorize(_lr)
```

```python
def _polynomial(step, init=0.01, end=0.0005, power=1, steps=100):
    step = min(step, steps)
    return ((init - end) * (1 - step / steps)**power) + end

poly = np.vectorize(_polynomial)
```

```python
x = np.arange(0, 500)
y = exp(x, decay=0.9, steps=10)
sns.lineplot(x=x, y=y)
```

```python
x = np.arange(0, 500)
y = poly(x, power=4, steps=200)
sns.lineplot(x=x, y=y)
```

```python
from keras.optimizers.schedules.learning_rate_schedule import PolynomialDecay
```

```python
poly_obj = PolynomialDecay(initial_learning_rate=0.01, decay_steps=100, power=1.5, cycle=True)
sns.lineplot(x=x, y=poly_obj(x))
```

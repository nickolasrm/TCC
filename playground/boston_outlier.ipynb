```python
from keras.datasets import boston_housing
```

```python
train, test = boston_housing.load_data()
```

```python
import seaborn as sns
import pandas as pd
```

```python
df = pd.DataFrame(train[0])
df = pd.melt(df, value_vars=df.columns)
df
```

```python
grid = sns.FacetGrid(df, col="variable", sharex=False, sharey=False)
grid.map(sns.histplot, "value", bins=10)
```
